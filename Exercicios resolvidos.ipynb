## **4. Funções**

As funções são o principal e mais importante método de organização e reutilização de código em Python

Como regra geral, se você prevê a necessidade de repetir o mesmo código ou um código muito semelhante várias vezes, pode valer a pena escrever uma função reutilizável.

As funções também podem ajudá-lo a tornar seu código mais legível nomeando um conjunto de instruções **Python**.

A função é declarada com a palavra-chave $def$ e retorna. Isso é feito usando a palavra-chave $return$:

```
def my_function(x, y, z=1.5):
  if z > 1:
    return z * (x + y)
  else:
    return z / (x + y)
```

### **4.2 Funções Anônimas (lambdas - $\lambda$)**

**Python** suporta o que chamamos de **funções anônimas ou lambdas** - uma maneira de escrever uma **função que consiste em uma única instrução cujo resultado é um valor de retorno**. Eles são definidos com a palavra reservada lambda, que não significa nada além de dizer "estamos declarando uma função anônima":

````python
def short_function(x):
  return x**2
````

````python
short_function(4)
````

````python
equiv_anon = lambda x: x ** 2
````

````python
#alguma coisa aqui
#...
equiv_anon(4)
````

Como em **Python** tudo é objeto, uma vantagem da função lâmbda está abaixo:

````python
def apply_to_list(alguma_lista, f):
  return [f(x) for x in alguma_lista]
````

````python
ints = [1, 3, 0, 60]

apply_to_list(ints, lambda x: x ** 2)
````

## **5. Tratamento de Exceções**

Lidar com erros ou exceções no **Python** é uma parte importante da construção de programas robustos. Em aplicativos de análise de dados, muitas funções funcionam apenas com determinados tipos de entrada. Por exemplo, a função float do Python pode converter uma string em vários
flutua, mas falha com um ValueError se os dados de entrada forem insuficientes:

````python
float('1.23231')
````

````python
float('1')
````

````python
float('teste')
````

````python
try:
  float('teste')
except:
  print('ops, deu erro')
````

````python
def cast_toFloat(objeto):
  try:
    float(objeto)
    return 'deu certo'
  except:
    return 'o objeto "{0}" não pode ser convertido em float'.format(objeto)
````

````python
print(cast_toFloat('1.234'))
print(cast_toFloat('1'))
print(cast_toFloat('teste'))
````

**Refinando os tipos de Erro**

Olha novamente a seguinte saída de erro:

````python
float('teste')
````

Agora vamos **refinar** o tratamento de erro para "pegar" apenas erros do tipo valor:

````python
def cast_toFloat(objeto):
  try:
    float(objeto)
    return 'deu certo'
  except ValueError:
    return 'o objeto "{0}" não pode ser convertido em float'.format(objeto)
````

````python
cast_toFloat('teste')
````

No entanto, veja a próxima instrução:

````python
cast_toFloat(1,234)
````

Leia: _**TypeError**: float() argument must be a string or a number, not 'tuple'_

Notou que são erros de tipo? Quando refinamos os erros, tratamos aquilo que é conhecido e deixamos o sistema "parar" (sim, isso pode ser uma boa estratégia), em erros desconhecidos. 

Uma possível incrementação neste tratamento, podemos ver abaixo:

````python
def cast_toFloat(objeto):
  try:
    float(objeto)
    return 'deu certo'
  except ValueError:
    return 'o objeto "{0}" não pode ser convertido em float'.format(objeto)
  except TypeError:
    return 'o objeto "{0}" não pode ser convertido em float erro de tipo'.format(objeto)
  except:
    return 'sei q deu, mas não sei o tipo'
````

````python
cast_toFloat('teste')
````

````python
cast_toFloat((1,234))
````

Em alguns casos, talvez você não queira suprimir uma exceção, mas queira que algum código seja executado, independentemente de o código no bloco try ter sido bem-sucedido ou não.

Um caso de uso seria quando você abre uma conexão com o banco de dados, e antes de fechá-lo (o que é uma excelente prática), seu código apresentou um erro. Assim, mesmo que você dê retorno do erro ao usuário (ou sistema), você precisa fechar a conexão com o banco de dados.

Para fazer isso, utilize $finally$:

````python
def manipula_dados(x):
  try:
    print('abre banco de dados')
    print('faz alguma coisa, como por exemplo tentar realizar um cast')
    a = float(x)
    #faz outras coisas
    print('deu tudo certo')
    return "mensagem de retorno: deu certo"
  except:
    print('tratamento de exceção')
    return "mensagem de retorno: deu erro, mas fica tranquilo que o BD está fechado"
  finally:
    print('fecha o banco de dados')
````

````python
manipula_dados(1)
````

````python
manipula_dados('teste')
````
