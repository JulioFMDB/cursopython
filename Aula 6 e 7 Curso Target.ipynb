**Enumerate**:

Imagina que você quer criar um dicionário (assunto que veremos em breve), para controlar os índices atuais dos elementos de uma _lista_ ou _tupla_. Uma maneira de se fazer é conforme abaixo:

### **3.4. List Comprehensions**

Caros alunos, agora vamos falar de um dos recursos mais interessantes do **Python**. Este recurso permite que você crie uma nova lista de modo conciso, filtrando os elementos de uma coleção, transformando os elementos ao passa o filtro, com uma expressão concisa. As **List Comprehensions** assumem o seguinte formato básico:

_`[expr for val in collection if condition]`_

Isso é equivalente ao laço a seguir:

```
result = []

for val in collection:
  if condition:
    result.append(expr)
```

A condição do filtro pode ser omitida, restando somente a expressão. Veja um exemplo:

````python
strings_a = 'Meu nome é Diego e eu sou um data scientist.'.split()
strings_a
````

````python
strings_b = [x.upper() for x in strings_a]
strings_b
````

````python
strings_b = [x.upper() for x in strings_a if len(x) > 2]
strings_b
````

````python
string_c = []

for x in strings_a:
  if len(x) > 2:
    string_c.append(x.upper())

string_c
````
